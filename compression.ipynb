```python
# Libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Read Image
img = mpimg.imread('images/chelsea.ppm')
(length, width, depth) = img.shape

plt.imshow(img)
print(f'Original image dimensions: Length: {length}, Width: {width}, Color Planes: {depth}')
```

```python
# Zero Padding the image
padding_window_size = 8
img_padded = np.pad(img, pad_width=[(0, padding_window_size - length%padding_window_size), (0, padding_window_size - width%padding_window_size), (0, 0)], mode='constant')
(length_padded, width_padded, depth_padded) = img_padded.shape

plt.imshow(img_padded)
print(f'Padded image dimensions: Length: {length_padded}, Width: {width_padded}, Color Planes: {depth_padded}')
```

```python
# Image conversion to YCbCr

rgb_to_yCbCr = np.array([[ 0.299, -0.168736,  0.5],
                          [0.587, -0.331264, -0.418688],
                          [ 0.114, 0.5, -0.081312]])

transformed_image = np.matmul(img_padded, rgb_to_yCbCr)

# Adjust encoding
transformed_image[:,:,1] += 128
transformed_image[:,:,2] += 128
```
